## PGM for observer inference

```python
import daft
from matplotlib import rc
from matplotlib.colors import to_rgba
import matplotlib.pyplot as plt
```

```python
def plot_pgm(
    fontfamily="serif",
    fontsize=12,
    fontcolor='royalblue',
    edgecolor='royalblue',
    observed_shade='#DEECFC',
    shade_observed=True,
    observed_outer=True,
    nodes_to_plot = ("d", "w", "o", "pi", "gi", "gb", "a0"),
    edges_to_plot = (("gi", "pi"), ("gb", "pi"), ("o", "pi"),
                     ("w", "pi"), ("pi", "d"), ("w", "d"), ("a0", "pi"),
                     ("s", "a"), ("pi", "a"), ("s", "ns"), ("a", "ns"),
                     ("w", "ns")),
    observed = ("w", "gi"),
    custom_node_params=None,
):
    rc("font", family=fontfamily, size=fontsize)
    rc("text", usetex=True)
    rc('text.latex', preamble=r'\usepackage{amsmath}') #for \text command


    defaultplot_params = {'ec': edgecolor}
    defaultedge_params = {"ec": edgecolor, "fc": edgecolor}
    defaultlabel_params = {'color': fontcolor, 'ha': 'center', 'va': 'center'}
    
    custom_node_params = custom_node_params if custom_node_params is not None else {}
    
    if observed_outer:
        pgm = daft.PGM(observed_style="outer")
    else:
        pgm = daft.PGM()

    node_params = {
        'd': dict( 
            node="d", 
            content=r"~\\$~~~~~~\,~\,\mbox{\huge\(D\)}$\\[-4pt]Demonstration", 
            x=4.0,
            y=1,
            aspect=2.2,
            scale=1.4,
            fontsize=10
        ),
        's': dict( 
            node="s", 
            content=r"Unpeeled", 
            x=2.5+.5,
            y=.7,
            aspect=1.9,
            scale=1.2,
            fontsize=12
        ),
        'ns': dict( 
            node="ns", 
            content=r"Peeled", 
            x=4.5+.5,
            y=.7,
            aspect=1.5,
            scale=1.2,
            fontsize=12
        ),
        'a': dict( 
            node="a", 
            content=r"Action", 
            x=3.5+.5,
            y=1.2,
            aspect=1.5,
            scale=1.2,
            fontsize=12
        ),
        'w': dict(
            node="w", 
            content=r"~\\$~~\,\mbox{\huge\(W\)}$\\World",
            x=4.2,
            y=3.0,
            aspect=1.4,
            scale=1.4,
            fontsize=12
        ),
        'o': dict(
            node="o",
            content=r"~\\$~~~\mbox{\huge\(\mathcal{O}_L\)}$\\Observer\\[-4pt]$~~~$Model",
            x=1.8,
            y=1.5,
            aspect=1.4,
            scale=1.55,
            fontsize=10
        ),
        "pi": dict(
            node="pi",
            content=r"$\!\!\!\!\!\!\;\!\mbox{\Huge\(\pi\)}$\\Plan",
            x=3.2,
            y=2,
            aspect=1.1,
            scale=1.4,
            fontsize=12
        ),
        "a0": dict(
            node="a0",
            content=r"$\!\!\!\mbox{\large\(\bar{A}\)}$\\[-3pt]Action\\[-3pt]$\,~\,~$Bias",
            x=3.2,
            y=3.2,
            aspect=1.1,
            scale=1.3,
            fontsize=10
        ),
        "gb": dict(
            node="gb",
            content=r"$~~\mbox{\huge\(G_B\)}$\\[-1pt]Belief-directed\\[-2pt]$~~~~~~~$Goal\vspace{-5pt}",
            x=1.4,
            y=2.27,
            aspect=2.,
            scale=1.5,
            fontsize=10
        ),
        "gi": dict(
            node="gi",
            content=r"~\\$~~~~~~\mbox{\huge\(G_I\)}$\\[-2pt]Instrumental\\[-2pt]$~~~~~~$Goal",
            x=2.05,
            y=3.,
            aspect=2.,
            scale=1.5,
            fontsize=10
        ),
    }
    for n in nodes_to_plot:
        pgm.add_node(**{
            **node_params[n], 
            "plot_params": {
                 **defaultplot_params,
                 **({'fc': observed_shade} if n in observed and shade_observed else {})
            },
            'label_params': {**defaultlabel_params},
            **custom_node_params.get(n, {})
        })
    for s, e in edges_to_plot:
        if s not in nodes_to_plot or e not in nodes_to_plot:
            continue
        pgm.add_edge(s, e, plot_params=defaultedge_params)

    img = pgm.render(dpi=300)
    return img
```

```python
plot_pgm(
    fontfamily="serif",
    fontsize=12,
    fontcolor='royalblue',
    edgecolor='royalblue',
    observed_shade='#DEECFC',
    shade_observed=True,
    observed_outer=True,
    nodes_to_plot = ("d", "w", "pi", "gi"),
    edges_to_plot = (("gi", "pi"), ("gb", "pi"), ("o", "pi"),
                     ("w", "pi"), ("pi", "d"), ("w", "d"), ("a0", "pi")),
    observed = ("w", "gi"),
)
```
