```python
%matplotlib inline

import json
import ast
import csv
from itertools import product
import matplotlib.pyplot as plt

import string
import random
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
responses = pd.read_pickle("../data/exp2b-responses.pd.pkl")
```

```python
random.seed(129347)
randstr = lambda : "".join(random.choice(string.ascii_letters) for i in range(15))
wids = sorted(list(responses.workerid.unique()))
wids = {w: randstr() for w in wids}
```

```python
responses['learner_id'] = responses['workerid'].apply(lambda w: wids[w])
```

```python
del responses['workerid']
```

```python
responses.to_csv("../data/exp2b-responses.csv")
```
