```python
%matplotlib inline

import json
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
responses = pd.read_pickle('../data/exp1L-trials.pd.pkl')
rf_order = ['ooo', 'oox', 'oxo', 'xoo', 'oxx', 'xox', 'xxo', 'xxx']
inst_order = ['doing', 'showing']
```

```python
g = sns.barplot(
                data=responses, 
                x='teacher_inst',
                ci=95,
                y='correct', hue='learner_inst',
                order=['doing', 'showing'],
                hue_order=['doing', 'showing'],
               capsize=.05,
               errwidth=2)
plt.ylim(.6, 1.05)
```
